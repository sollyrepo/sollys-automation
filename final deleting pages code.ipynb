```python
import glob 
import shutil 
import os
  
# Returns a list of names in list files. 
print("Using glob.glob()") 
files = glob.glob('C:/Users/DELL/Downloads/Saranya Check/**/*.PDF',recursive = True)
print(files)
#src = '/Users/pavithrathangarj/maathesh/sara/pdf_reader/testingpdf' 
dst =  'C:/sollys_automation_project/hand_written_files/'               
for file in files: 
    print(file) 
    from PyPDF2 import PdfFileWriter, PdfFileReader
    pages_to_delete = [0,1,2] # page numbering starts from 0
    infile = PdfFileReader(file,  'rb')#, strict = False
    output = PdfFileWriter()

    for i in range(infile.getNumPages()):
        if i not in pages_to_delete:
            p = infile.getPage(i)
            output.addPage(p)

    #shutil.move(src,dst)
    with open(dst+os.path.basename(file), 'wb') as f:
        output.write(f)
        
        
       # with Path("/Users/pavithrathangarj/maathesh/sara/pdf_reader/chapter1.pdf").open(mode="wb") as output_file:
   # output.write(output_file)

 # It returns an iterator which will  
# be printed simultaneously. 
#print("\nUsing glob.iglob()") 
#for filename in glob.iglob('/home/geeks/Desktop/gfg/**/*.txt', 
# recursive = True): 
#   print(filename) 
#with open('newfile.pdf', 'wb') as f:
    #output.write(f)
```
